```python
## importing dependencies
import numpy as np
import tensorflow as tf
import vgg19_style
import utils
import matplotlib.pyplot as plt
import pickle
```

We define here the gram matrix which takes a vector as the input the vector $F^l_{NXM}$ and constructs the $NXN$ gram matrix.

Next we define our main optimize function which generates the image starting from a white noise by minimizing the content and style losses.

```python
def optimize():
	l = []
	x_list = []
	with tf.Session() as session:
		session.run(init)
		# print("starting at", "x:", session.run(x))
		for step in range(num_epoch):  
			session.run(train)

			# session.run(tf.assign(x,tf.clip_by_value(x,0,1)))
			l.append(session.run(total_loss))
			print('iter',step)
			print('loss:', l[step])
			if(step%300==0):
				x_list.append(session.run(x))
		# x_list.append(session.run(x))

	return l,x_list
```

```python


style_image = utils.load_image("./test_data/van.jpg")     # content image
content_image = np.float32(utils.load_image("./test_data/cat.jpg"))


image_c = np.float32(content_image.reshape((1, 224, 224, 3)))
image_s = np.float32(style_image.reshape((1, 224, 224, 3)))
```

```python
#Parameters

num_epoch = 901    #number of epochs for training the input image
alpha = 0.001    #weight for content representation
beta = 1         # weight for style representation
```

```python
x = tf.Variable(tf.random_uniform(
    [1,224,224,3],
    minval=0,
    maxval=1,
    dtype=tf.float32,
    seed=None,
    name=None
),name="x")

model_c = vgg19_style.Vgg19()      # loading the model which computes the layer activations for content image
model_c.build(image_c)

model_s = vgg19_style.Vgg19()      # loading the model which computes the layer activations for style image
model_s.build(image_s)

model_x = vgg19_style.Vgg19()      # loading the model which computes the layer activations for the input image
model_x.build(x)


loss_content = (1/2)*tf.reduce_sum(tf.square(model_c.conv4_2 -  model_x.conv4_2 ))   # Content loss at layer conv4_2


loss_style = 0

loss_style += 0.2*tf.reduce_sum(tf.square(utils.gram_matrix(model_s.conv1_1) -  utils.gram_matrix(model_x.conv1_1)))
loss_style += 0.2*tf.reduce_sum(tf.square(utils.gram_matrix(model_s.conv2_1) -  utils.gram_matrix(model_x.conv2_1)))
loss_style += 0.2*tf.reduce_sum(tf.square(utils.gram_matrix(model_s.conv3_1) -  utils.gram_matrix(model_x.conv3_1)))
loss_style += 0.2*tf.reduce_sum(tf.square(utils.gram_matrix(model_s.conv4_1) -  utils.gram_matrix(model_x.conv4_1)))
loss_style += 0.2*tf.reduce_sum(tf.square(utils.gram_matrix(model_s.conv5_1) -  utils.gram_matrix(model_x.conv5_1)))

total_loss = alpha*loss_content + beta*loss_style#+ loss_lagrange


optimizer = tf.train.AdamOptimizer(learning_rate=0.05)


train = optimizer.minimize(total_loss)
init = tf.initialize_all_variables()


l,x_list = optimize()


a = x_list[-1]

image = a.reshape(224, 224, 3)
plt.figure()
plt.imshow(image)
plt.show()
```

The above image clearly depicts the style tranfer of the 'Van gough Starry night' painting onto our cat image. Running the code for a larger number of iteratinos will result in further smoothening and better style transfer.
